## Análise da Produção

```python
# Gráfico da produção
sns.distplot(train.production)
```

```python
# Análise dos dados de produção
train.production.value_counts()
```

```python
# Media da produção
print("Mean: ", train.production.mean())
```

```python
# Análise de todos os graficos em relação a produção
sns.pairplot(data=train, hue='production')
```

```python
# Gráfico do mês da colheita em relação a produção
f, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x='harvest_month', y="production", data=train)
```

```python
# Gráfico do ano da colheita em relação a produção
f, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x='harvest_year', y="production", data=train)
```

```python
# Gráfico do ano da colheita em relação a produção
f, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x='age', y="production", data=train)
```

```python
# Gráfico de tipos com a produção
f, ax = plt.subplots(figsize=(12, 6))
sns.countplot(x="type", palette="tab10", data=train)
```
